```python
# Visualise 7

data2 = df.copy()
data2.date = pd.to_datetime(data2['date'])
data2 = data2.groupby('date').sum()
data2['7 days MA new cases'] = 0
data2['7 days MA new cases'] = data2['new_cases'].rolling(7).mean()
data2['7 days MA new deaths'] = 0
data2['7 days MA new deaths'] = data2['people_fully_vaccinated'].rolling(7).mean()
data2['7 days MA new fully vaccinated'] = 0
data2['7 days MA new fully vaccinated'] = data2['people_fully_vaccinated'].rolling(7).mean()

data2[['new_cases', '7 days MA new cases']].plot(figsize = (11, 5), alpha = 0.5)
plt.title('Timeline New cases in the World')
plt.xlabel('Date')
plt.ylabel('New cases')
```

```python
# Visualise 8

data2[['new_deaths', '7 days MA new deaths']].plot(figsize = (11, 5), alpha = 0.5)
plt.title('Timeline New Deaths in the World')
plt.xlabel('Date')
plt.ylabel('New deaths')
```

```python
# Visualise 9

data2[['people_fully_vaccinated', '7 days MA new fully vaccinated']].plot(figsize = (11, 5), alpha = 0.5)
plt.title('Timeline People Fully Vaccinated in the World')
plt.xlabel('Date')
plt.ylabel('People Fully Vaccinated')
```
